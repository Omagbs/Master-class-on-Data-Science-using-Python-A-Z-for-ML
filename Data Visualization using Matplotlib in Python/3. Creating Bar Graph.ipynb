# Creating Bar Graph

```python
import matplotlib.pyplot as plt

# initializing the data 
x = [1, 2, 3, 4, 5, 6]
y = [45, 67, 74, 23, 83, 89]

# plotting the graph
plt.bar(x, y)

# Creating a title
plt.title("Performance Report")

# labeling x and y
plt.xlabel("Semester Numbers")
plt.ylabel("Percentage of Marks")

plt.show()
```
