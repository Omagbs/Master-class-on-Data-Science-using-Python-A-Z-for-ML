# Creating Line Graph

```python
import matplotlib.pyplot as plt
```

initialize data...

```python
# initialize data
x = [1, 2, 3, 4, 5]
y = [85, 87, 83, 86, 85]

# plotting the data
plt.plot(x, y)
# adding title
plt.title("Performance Report")

# labelling the axis..
plt.ylabel("Percentage of Marks")
plt.xlabel("Semester number")

# showing the graph
plt.show()
```
