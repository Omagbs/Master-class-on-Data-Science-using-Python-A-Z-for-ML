# Creating 3D plot
3D plot involves plotting data points on the x, y, and z axes.

```python
import matplotlib.pyplot as plt

# first create the values for x, y, z

x = [1 , 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]

fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot3D(x, y, z)
```
