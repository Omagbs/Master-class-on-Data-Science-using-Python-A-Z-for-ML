# Pie Chart

```python
import matplotlib.pyplot as plt

# data to display on plots
x = [1, 2, 3, 4]
e = (0, 0.1, 0, 0)

# e stands for explode. the higher the value, the more cutout/seperate a slice of a pice is from the whole pie together

# this will plot a simple pie plot
plt.pie(x, explode = e)

# title of the plot
plt.title("Pie Chart")
plt.show()
```

You can see that the orange slice of the pie is seperated from the overall pie. This is because of the second element of the explode being equated to 0.1. 
The higher the value, the more seperate.
