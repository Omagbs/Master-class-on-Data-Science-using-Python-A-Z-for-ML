# Swarm Plot
- A swarm plor is a type of scatter plot that is used for representing categorical values
- We can use seaborn.swarmplot() to create such graphs.

### To import the required libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading our Dataset:

```python
df = sns.load_dataset("tips")
```

### Visualizing our data using swarmplot:
we plot the tip observations and amount by the days of the week

```python
sns.swarmplot(x = 'day', y = 'tip', data = df)
sns.set_style("whitegrid")
plt.show()
```

```python
### Visualizing tip observation:
sns.swarmplot(x = df["tip"])
sns.set_style("darkgrid")
plt.show()
```

### Colours can be customizable.
So for instance, say i want to represent Male with a certain colour and likewise, a different one for female:

```python
gender_palette = ["#A833FF", "#7FAF33"]
sns.swarmplot(x = 'day', y = 'tip', hue = 'sex', palette = gender_palette, data = df)
plt.show()
```
