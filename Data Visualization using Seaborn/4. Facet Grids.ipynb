# Facet Grids
Facet grid takes a dataframe as input and the names of the variables will form the rows, columns or hue dimensions of the grid. The variables should be categorical and the data at each point level of the variable will be used for a facet along that axis

## Drawing Facet Grid based on tip variable

### Importing the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading out Data sets:

```python
df = sns.load_dataset("tips")
```

### Plotting facet grid with Histogram:

```python
fg = sns.FacetGrid(df, col = 'time', row = 'sex')
fg = fg.map(plt.hist, 'tip', color = 'tomato')
```

### Plotting facet grid with Scatter plots and changing the colours around the points:


```python
fg = sns.FacetGrid(df, col = 'time', row = 'sex)
fg = fg.map(plt.scatter,'total_bill', 'tip', color = 'FloralWhite', edgecolors = 'HotPink')
```

```python
d = {'color' : ['r', 'b']}
fg = sns.FacetGrid(df, col = 'time', hue = 'sex', hue_kws  = d)
fg = fg.map(plt.scatter,'total_bill', 'tip', color = 'FloralWhite', edgecolors = 'HotPink')
```

### Plotting by Category:

```python
x = sns.FacetGrid(df, col = 'time', hue = 'sex')
x = x.map(plt.scatter, 'total_bill', 'tip')
x = x.add_legend()
```
