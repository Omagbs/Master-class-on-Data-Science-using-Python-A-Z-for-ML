# Violin Plots

- is a statiscal representation of numerical data.
- it is similat to box plot, with the addition of a rotated kernel density plot on each side.

### To import necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading our data sets:

```python
df = sns.load_dataset("tips")
```

### Plotting:

```python
sns.violinplot(x=df["tip"], palette = ['gold'])
plt.show()
```

```python
sns.violinplot(x = 'sex', y = 'total_bill', data = df)
```
